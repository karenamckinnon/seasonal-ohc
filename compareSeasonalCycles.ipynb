```python
def calc_dA(lat, lon):
    # Approx area of global rectilinear grid cells
    # Approach taken from ncl approximation for consistency
    
    r = 6371220. # radius of earth, meters
    nlat, nlon = np.shape(lat)[0], np.shape(lon)[0]
    
    rad = np.pi/180.
    rr = r*rad
    dlon = rr*(lon[1] - lon[0])
    dx = dlon*np.cos(lat*rad)
    
    dy = np.empty(np.shape(lat))
    dy[0] = rr*(lat[1] - lat[0])
    dy[1:-1] = rr*(lat[2:] - lat[1:-1])
    dy[-1] = rr*(lat[-1] - lat[-2])
    
    area_lat = np.repeat((dx*dy)[:, np.newaxis], nlon, axis = -1)
    
    area_total = np.sum(area_lat.flatten())
    
    return area_lat
```

```python
# separate indices that span more than 10 degrees of latitude
latList = list()
lonList = list()
for i in np.arange(0, numL):

    lons, lats = argoLon[label_indices[i][1]], argoLat[label_indices[i][0]]
    dlats = np.max(lats) - np.min(lats)
    if dlats > 10:
        nSections = np.ceil(dlats / 10)
        newLatRange = np.linspace(np.min(lats), np.max(lats), nSections)
        newLatRange[0] -= 1e-5
        newLatRange[-1] += 1e-5
        # pull out new sets of lats/lons for each latRange pair
        latList.append([lats[(lats > newLatRange[jj]) & (lats <= newLatRange[jj + 1])] for jj in np.arange(0, newLatRange.shape[0]-1, 1)])
        lonList.append([lons[(lats > newLatRange[jj]) & (lats <= newLatRange[jj + 1])] for jj in np.arange(0, newLatRange.shape[0]-1, 1)])
    else:
        latList.append(lats)
        lonList.append(lons)
        
```

```python
def flatten(x):
    result = []
    for el in x:
      if (type(el).__module__ == '__builtin__') and not isinstance(el, basestring):
        result.extend(flatten(el))
      else:
        result.append(el)
    return result
```

```python
latList = flatten(latList)
lonList = flatten(lonList)
nregions = len(latList)
print 'There are ' + str(nregions) + ' total regions after limiting latitude range'
```

```python
def ismember_rows(a, b):
    # http://stackoverflow.com/a/22705773/190597 (ashg)
    return np.nonzero(np.all(b == a[:,np.newaxis], axis=2))[1]
```

```python
regionMask = np.zeros(totalMask.shape, dtype = 'int').flatten()
lons, lats = np.meshgrid(argoLon, argoLat)
lons = lons.flatten()
lats = lats.flatten()
for i in np.arange(nregions):
    lonsUse, latsUse = lonList[i], latList[i]
    idx = ismember_rows(np.vstack((lonsUse, latsUse)).T, np.vstack((lons, lats)).T)
    regionMask[idx] = i + 1
regionMask = regionMask.reshape(totalMask.shape)
```

## Use CESM to infer covariance structure for seasonal cycle
For each set of lats/lons, calculate the *sum* of the heat content, and use well-correlated gridcells to predict it.

Correlation only calculated with respect to annual cycle, NOT interannual variability or trend

Only allow positive coefficients for prediction equation

```python
LON, LAT = np.meshgrid(cesmLon, cesmLat)
LONvec = np.ndarray.flatten(LON)
LATvec = np.ndarray.flatten(LAT)

# remove annual mean from each year in CESM -- only trying to predict the shape of the seasonal cycle 
annMean = np.mean(np.reshape(cesmHeatMap, 
                                         (nyrs, moPerYear, cesmHeatMap.shape[1], cesmHeatMap.shape[2])), 
                              axis = 1)
modelAnomVals = np.reshape(cesmHeatMap, (nyrs, moPerYear, cesmHeatMap.shape[1], cesmHeatMap.shape[2])) - annMean[:, np.newaxis, :, :]
modelAnomVals = np.reshape(modelAnomVals, cesmHeatMap.shape)

# same for Argo
dummy = np.reshape(np.ma.average(np.reshape(argoHeatMap, 
                                         (nyrs, moPerYear, argoHeatMap.shape[1], argoHeatMap.shape[2])), 
                              axis = 1), (nyrs, 1, argoHeatMap.shape[1], argoHeatMap.shape[2]))

argoanomVals = np.reshape(argoHeatMap, (nyrs, moPerYear, argoHeatMap.shape[1], argoHeatMap.shape[2])) - dummy
argoanomVals = np.reshape(argoanomVals, argoHeatMap.shape)
argoHeatVec = np.reshape(argoanomVals, (argoanomVals.shape[0], LATvec.shape[0]))
```

```python
## Load CESM output (vertically integrated heat content) for estimation of trend
nc_fid = Dataset(cesm, 'r')
heatContentMapFull = nc_fid.variables['heatContentMapFull'][:]
heatContentMap20000 = nc_fid.variables['heatContentMap'][:]
time = nc_fid.variables['time'][:]
lat = nc_fid.variables['lat'][:]
lon = nc_fid.variables['lon'][:]
nlat, nlon, ntime = lat.shape[0], lon.shape[0], time.shape[0]

## Load Argo data, and then mask CESM output to match Argo data
#nc_fid = Dataset(argoUnfilled, 'r')
#heatContentMapArgo = nc_fid.variables['heatContentMap'][:]
heatContentMapArgo = np.empty_like(argoHeatMap)
heatContentMapArgo[:] = argoHeatMap
heatContentMap20000Masked = np.ma.masked_where(np.ma.getmask(heatContentMapArgo), heatContentMap20000)
heatContentMap20000MaskedVec = heatContentMap20000Masked.reshape((ntime, nlat*nlon))
heatContentMap20000Vec = heatContentMap20000.reshape((ntime, nlat*nlon))
```

```python
from scipy import stats, optimize 
import os
# cut off for how correlated things need to be to be used as a predictor
rhoPrcCut = 99

saveFile = outputDir + obsName +'.ohc.infilled.with.' + modelName + '.rho.gt.' + str(rhoPrcCut) + 'incl.trends.npy'

# this calculation can take a long time if not already done
if os.path.isfile(saveFile):
    filledArrays = np.load(saveFile)
else:
    argoFill = np.empty((nregions, argoHeatVec.shape[0]))
    modelOrig = np.empty((nregions, argoHeatVec.shape[0]))
    modelFill = np.empty((nregions, argoHeatVec.shape[0]))
    modelFillAll = np.empty((nregions, argoHeatVec.shape[0]))
    modelOrigAll = np.empty((nregions, argoHeatVec.shape[0]))
    rhoCutoff = np.empty(nregions)
    nPredLocs = np.empty(nregions)
    
    for ct in np.arange(nregions):
    #for ct in np.arange(1):
        print ct
        # find indices of lats/lons
        latUse = latList[ct]
        lonUse = lonList[ct]
        idx = np.squeeze(np.array([np.where(((LATvec == latUse[jj]) & (LONvec == lonUse[jj]))) for jj in np.arange(len(latUse))]))
        
        # add up heat in each region without data
        if len(modelHeatVec[:, idx].shape) > 1:
            sumHeat = np.sum(modelHeatVec[:, idx], axis = 1)
            sumHeat2 = np.sum(heatContentMap20000Vec[:, idx], axis = 1)

        else:
            sumHeat = modelHeatVec[:, idx]
            sumHeat2 = heatContentMap20000Vec[:, idx]


        # find other points that are well correlated with the sum of heat
        rho0 = np.array([stats.pearsonr(sumHeat, np.ma.getdata(modelHeatVecMasked)[:, ii]) 
                        for ii in np.arange(modelHeatVecMasked.shape[1])])

        # pull out indicies in the top 1% of correlations
        rho = rho0[:, 0]
        rho[np.sum(np.ma.getmask(modelHeatVecMasked), axis = 0) > 0] = np.nan
        rhoCutoff = np.percentile (rho[~np.isnan(rho)], rhoPrcCut)
        predictLocs = np.where((rho > rhoCutoff))

        # normalize by 1e22 for a more stable estimator
        A = np.squeeze(np.ma.getdata(modelHeatVecMasked)[:, predictLocs])
        b = sumHeat
        x = optimize.nnls(A/1e22, b/1e22)
        nPredLocs[ct] = np.sum(x[0] > 0)
        
        # fill in missing CESM data for check
        modelOrig[ct, :] = b
        modelFill[ct, :] = np.dot(A, x[0])

        # use to fill in missing argo data
        Aargo = np.squeeze(np.ma.getdata(argoHeatVec)[:, predictLocs])

        if (Aargo == -999.).any() | (np.abs(Aargo) > 1e36).any():
            print "Missing values in argo. Abort!"
        argoFill[ct, :] = np.dot(Aargo, x[0])
        
        filledArrays = np.dstack([modelOrig, modelFill, argoFill])
    
    np.save(saveFile, filledArrays)
```
